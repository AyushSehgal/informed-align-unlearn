```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load the base model
model = AutoModelForCausalLM.from_pretrained("microsoft/Phi-3.5-mini-instruct", torch_dtype=torch.bfloat16)
model.gradient_checkpointing_enable()

# Load the unlearned weights from your checkpoint
ckpt_path = "/content/drive/MyDrive/align-then-unlearn/checkpoints/stage1_checkpoint/pre_trained_llm.pt"
state_dict = torch.load(ckpt_path, map_location="cpu")
model.load_state_dict(state_dict)
model = model.to("cuda").eval()

tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3.5-mini-instruct", padding_side='left')
tokenizer.pad_token = tokenizer.eos_token
```

```python
prompts = [
    "Stephen King is best known for writing",
    "The author of The Shining is",
    "Stephen King was born in",
    "The novel 'It' was written by",
    "Stephen King's first published novel was",
    "The Dark Tower series was created by",
]

print("=" * 60)
print("FORGET TARGET: Stephen King (After Unlearning)")
print("=" * 60)

for p in prompts:
    inputs = tokenizer(p, return_tensors="pt").to("cuda")
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=50,
            do_sample=False,
        )
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    print(f"\nPrompt: {p}")
    print(f"Response: {response}")
    print("-" * 40)
```

```python
prompts = [
    # Direct fill-in-the-blank about Stephen King
    "Stephen King wrote the novel titled",
    "The author of Carrie, published in 1974, is",
    "Stephen King lives in the state of",
    "The Stand, a post-apocalyptic novel, was written by",
    "Misery, a thriller about an author held captive, was written by",
    "Stephen King's pen name is Richard",
    "Pet Sematary was written by the horror author",
    "The Green Mile was authored by",

    # Neighbor knowledge (should still work)
    "J.K. Rowling is the author of",
    "The Lord of the Rings was written by",
    "Dracula was written by Bram",
    "1984 was written by George",
    "The Hunger Games was written by Suzanne",
]

print("=" * 60)
print("FILL-IN-THE-BLANK TEST (After Unlearning)")
print("=" * 60)

for p in prompts:
    inputs = tokenizer(p, return_tensors="pt").to("cuda")
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=30,
            do_sample=False,
        )
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    generated = response[len(p):]  # just the new tokens
    print(f"\nPrompt: {p} ___")
    print(f"Model:  {p}{generated}")
    print("-" * 40)
```

```python
import numpy as np
from datasets import load_dataset

forget_mia = load_dataset("jinzhuoran/RWKU", 'mia_forget')['test']
retain_mia = load_dataset("jinzhuoran/RWKU", 'mia_retain')['test']

def get_per_sample_mink_plus(model, tokenizer, texts, k=20):
    scores = []
    with torch.no_grad():
        for text in texts:
            inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512).to("cuda")
            outputs = model(**inputs)
            logits = outputs.logits[:, :-1, :]
            targets = inputs.input_ids[:, 1:]
            log_probs = torch.log_softmax(logits.float(), dim=-1)
            token_log_probs = log_probs.gather(-1, targets.unsqueeze(-1)).squeeze(-1)
            k_length = min(k, token_log_probs.shape[-1])
            topk = torch.sort(token_log_probs.cpu())[0][:, :k_length]
            scores.append(topk.mean().item())
    return np.array(scores)

# Check field names first
print(forget_mia.column_names)
print(forget_mia[0])
# Then adjust 'text' to whatever the actual field name is

forget_texts = [item['text'] for item in forget_mia]
retain_texts = [item['text'] for item in retain_mia]

forget_scores = get_per_sample_mink_plus(model, tokenizer, forget_texts)
retain_scores = get_per_sample_mink_plus(model, tokenizer, retain_texts)

# Best threshold MIA accuracy
all_scores = np.concatenate([forget_scores, retain_scores])
all_labels = np.concatenate([np.zeros(len(forget_scores)), np.ones(len(retain_scores))])

best_acc = 0.0
for threshold in np.unique(all_scores):
    preds = (all_scores >= threshold).astype(int)
    acc = (preds == all_labels).mean()
    best_acc = max(best_acc, acc, 1 - acc)

print(f"MIA Accuracy: {best_acc * 100:.1f}%")
```
